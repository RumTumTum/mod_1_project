```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
imdb_title_basics = pd.read_csv("./data/imdb.title.basics.csv")
```

```python
imdb_title_basics = imdb_title_basics.loc[imdb_title_basics['genres'].isnull() == False]
```

```python
imdb_title_basics.shape
```

```python
imdb_title_basics.genres = imdb_title_basics.genres.map(lambda x : str(x).split(","))
```

```python
imdb_title_basics.genres.value_counts().head(5)
```

```python
unique_genres = []

for i in imdb_title_basics.genres:
    for j in i:
        if j in unique_genres:
            continue
        else:
            unique_genres.append(j)
```

```python
unique_genres
```

```python
genre_dict = {}

for genre in unique_genres:
    genre_dict[genre] = []
```

```python
genre_dict
```

```python
imdb_title_basics.head()
```

```python
for genre in unique_genres:
    bool_list = []
    for movie in imdb_title_basics.genres:
        if genre in movie:
            bool_list.append('True')
        else:
            bool_list.append('False')
    genre_dict[genre] = bool_list
        
```

```python
genre_dict
```

```python
for genre in unique_genres:
    imdb_title_basics[genre] = genre_dict[genre]
```
