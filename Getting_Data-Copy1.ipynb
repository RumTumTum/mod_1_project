```python
import pandas as pd
import movie_database_methods as imd
import get_imdb
```

```python
imdb = get_imdb.get_imdb()
```

```python
tn = imd.tn_movie_budgets()
```

```python
combined = imd.movies_combined(tn,imdb)
```

```python
combined.head(5)
```

```python
unique_genres= ['Action','Crime','Drama','Biography','Comedy','Fantasy','Horror','Thriller','Adventure','Animation',
'Documentary','History','Mystery','Sci-Fi','Romance','Family','War','Music','Sport','Western','Musical']
```

```python
combined['profit'] = combined.domestic_gross - combined.production_budget
```

### Runtime vs. Profit per Genre

```python
for genre in unique_genres[0:20]:
   
    combined.loc[combined[genre] == 'True'].plot(kind = 'scatter',x = 'runtime_minutes', 
                                                 xlim = (0,240), y = 'profit', title = genre)
```

```python
for genre in unique_genres[20:]:
   
    combined.loc[combined[genre] == 'True'].plot(kind = 'scatter',x = 'runtime_minutes', 
                                                 xlim = (0,240), y = 'profit', title = genre)
```

### Runtime vs. ROI per Genre

```python
combined['roi'] = combined['profit']/combined['production_budget']
```
